```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/olympics_clean.csv")
df.head()


# 10 países con más medallas
df.groupby("country_name")["total_medals"] \
  .sum() \
  .sort_values(ascending=False) \
  .head(10)


# Evolución de medallas por año
df.groupby("year")["total_medals"].sum().plot(figsize=(10,5))
plt.title("Evolución de medallas por año")
plt.xlabel("Año")
plt.ylabel("Total de medallas")
plt.tight_layout()
plt.show()


# Deportes con mayor participación
df["discipline_title"].value_counts().head(10)


# Distribución por tipo de medalla - tabla
df["medal_type"].value_counts()


sns.boxplot(x=df["total_medals"])
plt.title("Distribución de medallas")
plt.tight_layout()
plt.show()


# 10 países con más medallas
top_countries = (
    df.groupby("country_name")["total_medals"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_countries.plot(
    kind="bar",
    figsize=(10,5),
    title="Top 10 países con más medallas"
)
plt.ylabel("Total de medallas")
plt.tight_layout()
plt.show()


# Gráfico: Top 10 deportes
top_sports = df["discipline_title"].value_counts().head(10)

top_sports.plot(
    kind="barh",
    figsize=(10,5),
    title="Top 10 deportes con mayor participación"
)
plt.xlabel("Número de registros")
plt.tight_layout()
plt.show()


# Gráfico: Distribución de tipos de medalla
medals = df["medal_type"].value_counts()

medals.plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6),
    title="Distribución de tipos de medalla"
)
plt.ylabel("")
plt.tight_layout()
plt.show()
```

## Conclusiones


- El análisis muestra una tendencia histórica de crecimiento en el número total de medallas otorgadas, asociada a la expansión de disciplinas y eventos olímpicos.
- La evolución temporal presenta variaciones entre ediciones, debido a la alternancia entre Juegos Olímpicos de Verano e Invierno.
- El total de medallas se concentra en un grupo reducido de países, liderados históricamente por Estados Unidos, lo que evidencia desigualdad en el rendimiento deportivo global.
- Atletismo y natación destacan como los deportes con mayor participación, debido a la diversidad de pruebas y categorías que presentan.
- La distribución de medallas es altamente asimétrica: la mayoría de los registros no obtienen medalla, mientras que solo un pequeño porcentaje alcanza oro, plata o bronce.
- La presencia de valores atípicos en la distribución de medallas refleja países o participaciones con rendimiento excepcional, y no representa errores en los datos
